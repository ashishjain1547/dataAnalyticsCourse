```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

Logistic regression is powerful widely used in industry tool. It can be seen a little bit tricky due to its cost function. However this pays off.

The cornerstone of logistic regression is logistic sigmoid function which is inverse of logit function.
Equation below is sigmoid function, where '-x' is the linear combination of weights we need to find out in order to separate our classes.\
The output of the sigmoid function is interpreted as the probability of a particular example belonging to class 1



```python
# Sigmoid function
def sigmoid (x):
    return 1.0 / (1.0 + np.exp(-x))

x=np.arange(-10,10,0.1)
f_x=sigmoid(x)
plt.plot(x,f_x)
plt.axvline(0.0,c='black',ls='--',lw=0.5)
plt.axhline(0.5,c='black',ls='--',lw=0.5)
plt.rcParams["figure.figsize"] = (14,7)
plt.show()
```

Logistic regression cost function



As an example of how does logistic regression cost function work lets consider outcome where we have Y equals to 1 and Y_pred equals from 0.5 to 1

```python
Y_pred=[0.5,0.6,0.7,0.8,0.9,0.999]

Y=1.0

for i in Y_pred:
    
    print('if Y_pred =',i,'and Y=1, cost =',-Y*np.log(i)-(1-Y)*np.log(1-i))
```

Derivative of logistic regression cost function



```python
# Toy dataset using sklearn make_blobs function

X,y=make_blobs(n_samples=100,n_features=2,centers=2,random_state=6,cluster_std=1.3)
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
class LogReg (object):
    
    def __init__ (self,n_iter,l_rate,random_state):
        
        self.n_iter=n_iter # number of itirations 
        self.l_rate=l_rate # learning rate
        self.random_state=random_state # random_state for weights generation
        
    def fit (self,X,y): # weights update process and cost monitoring 
        
        # for initial weights assignment I use numpy.random.RandomState which is a
        # container with number of methods for random numbers generation
        # drawn from a variety of probability distributions
        
        rgen=np.random.RandomState(self.random_state)
        
        # weights initialisation
        # normal (Gaussian) distribution is used with mean equals to 0.0 and std to 0.01 and size
        # corresponding to number of features + bias (intercept)
        
        self.w_=rgen.normal(loc=0.0,scale=0.01,size=X.shape[1]+1)
        
        self.cost_=[]
        
        for i in range(self.n_iter):
            
            output=(self.sigmoid(self.dot_product(X)))
            
            errors=(y-output) # error calculation
            
            self.w_[:2]+=self.l_rate*X.T.dot(errors) # weights update process
            
            self.w_[2]+=self.l_rate*errors.sum()
            
            cost = (-y.dot(np.log(output))-((1 - y).dot(np.log(1 - output)))) # cost function
            
            self.cost_.append(cost)
        
        return self
        
    def dot_product (self,X): 
        
        return np.dot(X,self.w_[:2])+self.w_[2]
        
    def sigmoid (self,z): # Sigmoid function
        
        return 1/(1+np.exp(-z))
        
    def predict (self,X): # Prediction function using np.where
        
        return np.where(self.sigmoid(self.dot_product(X))>=0.5,1,0)
        
```

While performing train-test split it is always a good idea to compare train and test samples means to be sure that there is no significant stat difference between samples.

If difference in sample means looks suspicious I prefer to apply t-tes. However in current case means are ok.

```python
X_train=X[:70]
X_test=X[70:]
y_train=y[:70]
y_test=y[70:]

print('mean_1:',X_train[:,0].mean(),X_test[:,0].mean())

print('mean_2:',X_train[:,1].mean(),X_test[:,1].mean())
```

```python
lr=LogReg(n_iter=10000,l_rate=0.001,random_state=1)
lr.fit(X_train,y_train)

y_pred=lr.predict(X_test)
print('Misclassifications:',(y_test!=y_pred).sum())
```

```python
# Ploting cost value vs number of iterations
plt.plot(np.arange(1,len(lr.cost_)+1,1),lr.cost_)
```

```python
# Finding slope and intercept for decission line ploting
slope=(0-(-lr.w_[2]/lr.w_[1]))/(-lr.w_[2]/lr.w_[0])
intercept=-lr.w_[2]/lr.w_[1]

print('slope:',slope,'intercept:',intercept)
```

```python
# Drawing decission line over test dataset
plt.scatter(X_test[:,0],X_test[:,1],c=y_test)
plt.plot(np.linspace(X_test[:,0].min(),X_test[:,0].max(),100),
        (np.linspace(X_test[:,0].min(),X_test[:,0].max(),100))*slope+intercept,'r--')

plt.show()
```
