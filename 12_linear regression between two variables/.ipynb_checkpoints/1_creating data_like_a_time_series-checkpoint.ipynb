```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.arange(100)

y = 0.4 * x + 3
```

```python
plt.plot(x, y, linestyle = 'dotted')
```

# Adding noise based on normal distribution

```python
mean, std, sample_size = 0, 5, 100 # mean and standard deviation

noise = np.random.normal(mean, std, sample_size)
```

```python
noise[:20]
```

```python
plt.hist(noise)
plt.show()
```

```python
x = np.arange(100)
y = (0.4 * x + 3) + noise
plt.plot(x, y, linestyle = 'dotted')
```

```python
y = [round(i, 2) for i in y]
df = pd.DataFrame({
    'x': x,
    'y': y
})
```

```python
df.to_csv('two_linearly_related_variables_with_noise.csv', index = False)
```
