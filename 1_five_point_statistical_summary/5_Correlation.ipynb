# Calculating correlation manually

```python
a = [random.randint(i, i + 20) for i in range(100)]
b = [random.randint(i*20, i*20 + 200) for i in range(100)]
```

```python
print(a)
```

## 1: Calculation of means

```python
N = len(a)
a_bar = st.mean(a)
b_bar = st.mean(b)
```

## 2: Calculation of standard deviation (for denominator)

```python
sigma_a = st.stdev(a)
sigma_b = st.stdev(b)
```

## 3: Differences from the respective means

```python
a_diff_from_mean = [item - a_bar for item in a]
```

```python
b_diff_from_mean = [item - b_bar for item in b]
```

## 4: Product of differences from the means

```python
product_of_differences_from_mean = [
    i[0] * i[1] 
    for i in zip(a_diff_from_mean, b_diff_from_mean)
]
```

## 5: Calculation of sum of product of differences from mean

```python
s = sum(product_of_differences_from_mean)
```

## 6: Last calculation for correlation

```python
corr = s / (N * sigma_a * sigma_b)
```
